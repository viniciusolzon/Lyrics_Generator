```python
linkin_park_df = df.loc[df['artist'] == 'Linkin Park Lyrics']
```

```python
linkin_park_df = linkin_park_df.reset_index()
```

```python
linkin_park_df.drop(['index', 'Unnamed: 0', 'link', 'artist', 'song_name'], axis='columns', inplace=True)
```

```python
linkin_park_df = linkin_park_df.rename(columns={'lyrics': 'lyric'})
```

```python
regex_pattern = r'\[\w+:\]'
# Filtrar as linhas que contÃªm palavras que correspondem ao padrÃ£o regex
linhas_filtradas = linkin_park_df[linkin_park_df['lyric'].str.contains(regex_pattern, regex=True)]
linhas_filtradas
```

```python
regex_pattern = r'\[\w+:\]'
regex_pattern = r'\[[^\]]+:\]'
# Filtrar as linhas que contÃªm palavras que correspondem ao padrÃ£o regex
linkin_park_df['lyric'] = linkin_park_df['lyric'].str.replace(regex_pattern, '', regex=True)
```

```python
# ExpressÃ£o regular para encontrar palavras que comeÃ§am com "[" e terminam com ":]"
regex_pattern = r'\[\w+:\]'

# Remover palavras que correspondem ao padrÃ£o regex e tambÃ©m quaisquer espaÃ§os extras
linkin_park_df['lyric'] = linkin_park_df['lyric'].str.replace(regex_pattern, '', regex=True).str.replace(r'\s+', ' ').str.strip()
```

```python
regex_pattern = r'\[\w+:\]'

# Filtrar as linhas que contÃªm palavras que correspondem ao padrÃ£o regex
linhas_filtradas = linkin_park_df[linkin_park_df['lyric'].str.contains(regex_pattern, regex=True)]
linhas_filtradas
```
